```python
df_april = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-apr14.csv')
df_may = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-may14.csv')
df_june = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-jun14.csv')
df_july = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-jul14.csv')
df_aug = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-aug14.csv')
df_sept = pd.read_csv(r'C:\Users\Navaneeth\Desktop\Python\uber-raw-data-sep14.csv')
dfs = [df_april, df_may, df_june, df_july, df_aug, df_sept]
uber_data = pd.concat(dfs, ignore_index=True)

```

```python
uber_data['Date/Time'] = pd.to_datetime(uber_data['Date/Time'], format="%m/%d/%Y %H:%M:%S")
uber_data['DayOfWeekNum'] = uber_data['Date/Time'].dt.dayofweek
uber_data['DayOfWeek'] = uber_data['Date/Time'].dt.day_name()
uber_data['MonthDayNum'] = uber_data['Date/Time'].dt.day
uber_data['HourOfDay'] = uber_data['Date/Time'].dt.hour
```

```python
uber_weekdays = uber_data.pivot_table(index=['DayOfWeek'],
                                  values='Base',
                                  aggfunc='count')
uber_weekdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Week Day');
```

```python
uber_data.groupby(['Base','MonthDayNum']).count().unstack('Base')['Date/Time'].plot(figsize = (8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Month by Dispatch Base')
```
